Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
z = [3,4,5,6,1,9,4]
c = ["r","b","g","y","r","b","g"]
s = [500,300,200,200,300,200]

plt.xlabel("X-axis",fontsize = 15)
plt.ylabel("Y-axis",fontsize = 15)
plt.title("Scatter Plot",fontsize = 20)

# plt.scatter(x, y,color=c, sizes=s, alpha=0.6, marker='*', edgecolor="black",cmap="viridis")

plt.scatter(x, y,color=c, sizes=s, alpha=0.6, marker='*', edgecolor="black")
plt.scatter(x, z,color="r", sizes=s, alpha=0.6, marker='*', edgecolor="black")
# t= plt.colorbar()

# t.set_label("Color Bar",fontsize = 15)
plt.show()
```
