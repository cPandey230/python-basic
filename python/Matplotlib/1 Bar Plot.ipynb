Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x = ["python" , "C" , "C++" , "Java"]
y = [85,70,60,82]
c = ["r","b","g","y"]
l= ["python" , "C" , "C++" , "Java"]

plt.xlabel("Language",fontsize = 15)
plt.ylabel("Popularity" ,fontsize = 15)
plt.title("Programming Language",fontsize = 20)

# plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="black",linewidth=2,linestyle =":",alpha=0.8)
plt.bar(x,y,label=l,color=c)
plt.legend()
plt.show()
```

```python
import numpy as np

x = ["python" , "C" , "C++" , "Java"]
y = [85,70,60,82]
z = [80,60,50,75]
p = np.arange(len(x))
width = 0.2
p1 =[j+width for j in p]


plt.xlabel("Language", fontsize = 15)
plt.ylabel("Popularity", fontsize = 15)
plt.title("Programming Language", fontsize = 20)


plt.bar(p,y, width, label="popularity", color="r",)
plt.bar(p1,z, width, label="popularity1", color="b")

plt.xticks(p+width/2,x,rotation =10)
plt.legend()
plt.show()
```

```python
import numpy as np

x = ["python" , "C" , "C++" , "Java"]
y = [85,70,60,82]
z = [80,60,50,75]
p = np.arange(len(x))
width = 0.2
p1 =[j+width for j in p]


plt.xlabel("Language", fontsize = 15)
plt.ylabel("Popularity", fontsize = 15)
plt.title("Programming Language", fontsize = 20)


plt.barh(p,y, width, label="popularity", color="r",)
plt.barh(p1,z, width, label="popularity1", color="b")

plt.yticks(p+width/2,x,rotation =10)
plt.legend()
plt.show()
```
