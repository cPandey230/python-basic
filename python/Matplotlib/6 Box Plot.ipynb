BOX PLOT OR WHISKER PLOT

```python
import matplotlib.pyplot as plt
```

```python
x =[10,20,30,40,50,60,70]

# plt.boxplot(x,notch=False,vert=True,widths=0.2,tick_labels=["python"],patch_artist=False,
#            showmeans=True,whis=2,filerprops=dict(markedgecolor="y"))
plt.boxplot(x,notch=False,vert=True,widths=0.2,tick_labels=["python"],patch_artist=False,
           showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r")
           )
plt.show()
```

```python
x =[10,20,30,40,50,60,70]
x1=[10,20,40,50,30,60,90]
x2 = [20,30,60,50,40,80,100]
y= [x,x1,x2]


plt.boxplot(y,widths=0.2,tick_labels=["python","java","c++"])
plt.show()
```
