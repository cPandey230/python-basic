Filling Between Area of Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y =  np.array([1,2,3,4,5])

plt.title("Filling Between Plot")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")

plt.plot(x,y,color="r")
plt.fill_between(x,y,color="g",where= (x>2) &  (x <=4),alpha =0.6)
# plt.fill_between(x=[2,3],y1=2,y2=4,color="g")
plt.show()
```
