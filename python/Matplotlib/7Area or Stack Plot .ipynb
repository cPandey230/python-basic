Stack Plot or Area Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l= ["Area1","Area2","Area3"]
c= ["r","g","m"]

plt.title("Python")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.stackplot(x,area1,area2,area3,labels=l,colors=c,baseline="zero")
plt.legend(loc=2)
plt.show()
```
