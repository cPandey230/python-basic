PIE CHART

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]
ex= [0.1,0.0,0.0,0.0]
c = ["pink","blue","orange","red"]

plt.title("Programming Language",fontsize = 20)

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=0.9,labeldistance=1.1,startangle=0,
        textprops={"fontsize" : 11},counterclock=False,wedgeprops= {"linewidth" : 0.3,"edgecolor" : "black"},rotatelabels=False ) 
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
x = [10,20,30,40]
x1 = [15,25,35,45]
y = ["C","C++","Java","Python"]
ex= [0.1,0.0,0.0,0.0]
c = ["pink","blue","orange","red"]

plt.title("Programming Language",fontsize = 20)

plt.pie(x,labels=y,radius=1,) 
plt.pie(x1,labels=y,radius=0.5) 
plt.show()
```

```python
x = [10,20,30,40]
x1 = [15,25,35,45]
y = ["C","C++","Java","Python"]

plt.title("Programming Language",fontsize = 20)

plt.pie(x,labels=y,radius=1.5) 
plt.pie([1],colors="w") 
plt.show()
```

```python
x = [10,20,30,40]
x1 = [15,25,35,45]
y = ["C","C++","Java","Python"]
cr = plt.Circle(xy=(0,0),radius=1,facecolor="white")


plt.title("Programming Language",fontsize = 20)

plt.pie(x,labels=y,radius=1.5) 
plt.gca().add_artist(cr)
plt.show()
```
