STEM PLOT

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

# plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=0,basefmt="g",label="Python",use_line_collection=False)
plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=0,basefmt="g",label="Python",orientation="vertical")

plt.legend()
plt.show()
```
