```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
df = pd.read_excel(r"C:\Users\chandan\Desktop\python\z1DataCleaning\TestDataCleaning\cleaned_data.xlsx")
df
```

```python
df["Registration_Date"] = pd.to_datetime(df["Registration_Date"])
df
```

```python
# Trim whitespace
df["ID"] = df["ID"].str.strip()
df["Email"] = df["Email"].str.strip()
df["Text"] = df["Text"].str.strip()

df["Email"] = df["Email"].str.lower()
df["Text"] = df["Text"].str.lower()
df
```

```python
df['Text'] = df['Text'].str.replace(r'@(a-zA-Z0-9\s)','',regex=True)
df
```

```python
df["Domain"] = df["Email"].str.extract(r"@([a-zA-Z0-9.-]+)")
df
```

```python
# Normalize phone numbers to a common format
df["Phone"] = df["Phone"].str.replace(
    r"\D", "", regex=True
)  # Remove all non-numeric characters
df["Phone"] = df["Phone"].apply(
    lambda x: (
        re.sub(r"(\d{3})(\d{3})(\d{4})", r"\1-\2-\3", x)
        if pd.notnull(x) and len(x) == 10
        else x
    )
)
df
```

```python
#  Handle missing values
df['ID'] = df['ID'].fillna('unknown')
df['Email'] = df['Email'].fillna('missing@example.com')
df['Text'] = df['Text'].fillna('No text provided')
df['Phone'] = df['Phone'].fillna('000-000-0000')
df['Age'] = df['Age'].fillna(df['Age'].mean())  # Fill missing Age with the mean age
df
```
