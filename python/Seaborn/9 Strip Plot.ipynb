Strip Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
# sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="Dark2",linewidth=1,edgecolor="y",jitter=10,)
m={
    "Male":"o",
   "Female":">",
}
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="Dark2",size=5,alpha=0.9,)
plt.title("Strip Plot")
plt.show()
```
