Count plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex" , data=var , hue = "smoker",color="r",alpha=0.9,saturation=0.3)
plt.show()
```

```python
sns.barplot(x = "sex",y="size",data=var,hue="smoker")
```
