Violin Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
color = ["o","g","r","b"]
sns.violinplot(x="day",y="total_bill",data=var,hue="time",linewidth=2,palette="Dark2")
plt.title("Vioin Plot")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"])
plt.title("Vioin Plot")
plt.show()
```

```python
sns.violinplot(y="day",x="total_bill",data=var,hue="sex",split=True,inner="quartz")
plt.title("Vioin Plot")
plt.show()
```

```python
sns.violinplot(x=var["total_bill"])
plt.show()
```
