Styling Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python

sns.barplot(x="day",y="total_bill",data=var)

sns.set_context("paper",font_scale=217)
sns.despine()

plt.title("Bar Plot",fontsize=15)
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(6,6))
sns.barplot(x="day",y="total_bill",data=var,hue="sex",palette="plasma")

sns.despine()

plt.title("Bar Plot",fontsize=15)
plt.show()
```
