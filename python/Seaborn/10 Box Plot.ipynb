Box Plt

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",palette="Dark2",order=["Fri","Sun","Thur","Sat"],showmeans=True,
           meanprops={"marker": "+","markeredgecolor" : "black"},linewidth=1)
plt.title("Box Plot")
plt.show()
```

```python
# for horizontal both data should be numerical or you can swap the axis
sns.set(style="darkgrid")
sns.boxplot(y="day",x="total_bill",data=var,hue="sex",palette="plasma",order=["Fri","Sun","Thur","Sat"],showmeans=True,
           meanprops={"marker": "+","markeredgecolor" : "black"},linewidth=1)
plt.title("Box Plot")
plt.show()
```

```python
# for horizontal both data should be numerical or you can swap the axis
sns.set(style="darkgrid")
sns.boxplot(data=var,palette="plasma",orient="h" )
plt.title("Box Plot")
plt.show()
```

```python
sns.set(style="darkgrid")
sns.boxplot(y=var["total_bill"])
plt.title("Box Plot")
plt.show()
```
