```python
import pandas as pd
var = pd.DataFrame(
    {
        "A" : [1,2,3,4],
        "B" : [5,6,7,8]
    }
)
var
```

```python
var["Sum "] = var["A"] + var["B"]
var["Sub "] = var["A"] - var["B"]
var["Mul "] = var["A"] * var["B"]
var["div "] = var["A"] / var["B"]
var
```
