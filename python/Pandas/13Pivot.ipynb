pivot()

```python
import pandas as pd
```

```python
var = pd.DataFrame(
    {
        "Days" : [1,2,3,4,5,6],
        "Name" : ["a", "b" ,"c","a", "b" ,"c" ],
        "English" : [10,12,13,14,15,15],
        "Maths" : [16,17,18,19,20,21],
    }
)
var
```

```python
var.pivot(index = "Days" , columns = "Name")
```

```python
var.pivot(index = "Days" , columns = "Name",values ="English")
```

```python
var = pd.DataFrame(
    {
        "Days" : [1,1,1,1,2,2],
        "Name" : ["a", "b" ,"a","a", "b" ,"b" ],
        "English" : [10,12,13,14,15,15],
        "Maths" : [16,17,18,19,20,21],
    }
)
var
```

```python
var.pivot_table(index="Name",columns="Days").mean
```
