```python
var1 = pd.DataFrame(
    {
        "A" : [1,2,3,4],
        "B" : [11,12,13,14],
    }
)
var2= pd.DataFrame(
    {
        "A" : [1,2,3,5],
        "C" : [21,22,23,24],
    }
)
pd.merge(var1,var2, on= "A")
```

```python
var1 = pd.DataFrame(
    {
        "A" : [1,2,3,4],
        "B" : [11,12,13,14],
    }
)
var2= pd.DataFrame(
    {
        "A" : [1,2,3,5],
        "C" : [21,22,23,24],
    }
)
```

```python
pd.merge(var1,var2, how="left")
```

```python
pd.merge(var1,var2, how="right")
```

```python
pd.merge(var1,var2, how="outer")
```

```python
pd.merge(var1,var2, how="outer" , indicator=True)
```

```python
var1 = pd.DataFrame(
    {
        "A" : [1,2,3,4],
        "B" : [11,12,13,14],
    }
)
var2= pd.DataFrame(
    {
        "A" : [1,2,3,5],
        "B" : [21,22,23,24],
    }
)
pd.merge(var1,var2,left_index=True,right_index=True,suffixes =["_Data","_Value"] )
```
