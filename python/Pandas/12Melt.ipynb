Melt()

```python
import pandas as pd
```

```python
var = pd.DataFrame(
    {
        "Days" : [1,2,3,4,5,6],
        "English" : [10,12,13,14,15,15],
        "Maths" : [16,17,18,19,20,21],
    }
)
var
```

```python
pd.melt(var)
```

```python
pd.melt(var,id_vars=["English"])
```

```python
pd.melt(var,id_vars=["English"],var_name="python",value_name ="wscube")
```
