```python
import pandas as pd
```

```python
s1 = pd.Series([1,2,3,4])
s2 = pd.Series([11,23,31,41])

pd.concat([s1,s2])
```

```python
var1 = pd.DataFrame(
    {
        "A" : [1,2,3,4],
        "B" : [11,12,13,14],
    }
)
var2= pd.DataFrame(
    {
        "A" : [1,2,3,4],
        "C" : [21,22,23,24],
    }
)
pd.concat([var1,var2])
# pd.concat([var1,var2],"axis=0")
```

```python
pd.concat([var1,var2],axis=1)
```

```python
var1 = pd.DataFrame(
    {
        "A" : [1,2,3,4],
        "B" : [11,12,13,14],
    }
)
var2= pd.DataFrame(
    {
        "A" : [1,2],
        "C" : [21,22],
    }
)
pd.concat([var1,var2],axis=1)
```

```python
pd.concat([var1,var2],axis=1,join= "outer")
```

```python
pd.concat([var1,var2],axis=1,join= "inner")
```

```python
pd.concat([var1,var2],axis=0,keys=["D1","D2"])
```

```python
pd.concat([var1,var2],axis=1,keys=["D1","D2"])
```
